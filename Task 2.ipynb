```python
import turtle

def draw_rectangle(color, width, height):
    turtle.begin_fill()
    turtle.fillcolor(color)
    for _ in range(2):
        turtle.forward(width)
        turtle.right(90)
        turtle.forward(height)
        turtle.right(90)
    turtle.end_fill()

def draw_ashoka_chakra(radius):
    turtle.penup()
    turtle.goto(0, -radius)
    turtle.pendown()
    turtle.color("navy")
    turtle.width(2)
    turtle.circle(radius)
    
    for _ in range(24):
        turtle.penup()
        turtle.goto(0, 0)
        turtle.pendown()
        turtle.forward(radius)
        turtle.right(15)

def draw_indian_flag(flag_width, flag_height, chakra_diameter):
    saffron_height = white_height = green_height = flag_height / 3
    
    turtle.speed(10)
    turtle.penup()
    turtle.goto(-flag_width / 2, flag_height / 2)
    turtle.pendown()

    draw_rectangle("orange", flag_width, saffron_height)

    turtle.penup()
    turtle.goto(-flag_width / 2, saffron_height / 2)
    turtle.pendown()

    draw_rectangle("white", flag_width, white_height)

    turtle.penup()
    turtle.goto(-flag_width / 2, -flag_height / 6)
    turtle.pendown()

    draw_rectangle("green", flag_width, green_height)

    turtle.penup()
    turtle.goto(0, 0)
    turtle.pendown()
    draw_ashoka_chakra(chakra_diameter / 2)

    turtle.hideturtle()
    turtle.done()

flag_width = float(input("Enter the width of the flag: "))
flag_height = float(input("Enter the height of the flag: "))
chakra_diameter = float(input("Enter the diameter of the Ashoka Chakra: "))

draw_indian_flag(flag_width, flag_height, chakra_diameter)
```
